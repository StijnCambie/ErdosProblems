```python
def delta1(n,m):
    T=0;
    v=range(n+1,m);
    for k in range(1,len(v)+1):
        som=0;
        for j in Subsets(v,k):
            som+=1/LCM(j);
        T+=(-1)^(k+1)*som*k;
    return T
            
def Unimodal(a):
    M=max(a);
    for i in range(1,len(a)-1):
        if max(a[:i])<M and a[i]<a[i-1]:
            return False
        if max(a[:i])==M and a[i]<a[i+1]:
            return False
    return True
```

```python
for n in range(2,21):
    L=[]
    for k in range(2,16):
        L.append(round(delta1(n,n+k),4))
    print(L)
    if Unimodal(L):
        print(str(n)+"unimodal at the start")
```
